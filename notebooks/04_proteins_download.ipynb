```python
pdb_ids = ['6GJ7', '7AVL', '6GJ6', '6ZIR', '6ZIZ', '6ZJ0', '6ZL3', '7UKR', '7UKS']

output_dir = "structures_raw"
os.makedirs(output_dir, exist_ok=True)

formats = ['pdb', 'cif']

print(f"Starting direct download of {len(pdb_ids)} structures in PDB and mmCIF formats...")

for pdb_id in pdb_ids:
    for fmt in formats:
        url = f"https://files.rcsb.org/download/{pdb_id}.{fmt}"
        save_path = os.path.join(output_dir, f"{pdb_id}.{fmt}")

        try:
            response = requests.get(url)
            if response.status_code == 200:
                with open(save_path, 'wb') as f:
                    f.write(response.content)
                print(f"Downloaded: {pdb_id}.{fmt}")
            else:
                print(f"Failed: {pdb_id}.{fmt} (Status code: {response.status_code})")
        except Exception as e:
            print(f"Error downloading {pdb_id}.{fmt}: {e}")

print(f"\nDone! All files are in: {os.path.abspath(output_dir)}")
```
